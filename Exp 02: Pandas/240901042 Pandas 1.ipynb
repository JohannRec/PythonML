```python
import pandas as pd
#empty dataframe
df = pd.DataFrame()
print(df)
```

```python
emp = pd.Series(['Parker', 'John', 'Smith', 'William'])
id = pd.Series([102, 107, 109, 114])

frame = {'Emp':emp, 'ID':id}
result = pd.DataFrame(frame)
print("Series to Data frame")
print(result)
```

```python
print ("Extrating one column:\n")
print (result['Emp'])
```

```python
print("Adding new column:\n")
result['Age'] = pd.Series([35, 24, 40, 38])
print(result)
```

```python
print("Deleting one column:\n")
del result['Age']
print(result)
```

```python
print("Extracting the third row:\n")
print(result.loc[2])
```

```python
print("Slice rows:\n", result[1:3])
```

```python
d2 = pd.DataFrame([['Dale', 123], ['Mark', 143]], columns = ['Emp', 'ID'])
print("Adding new row values:\n", pd.concat([result, d2]))
```

```python
print("Deleting particular row:\n", result.drop(1))
```
