#### Aim: 
To perform array slicing and indexing operation towards demonstrating various array joining techniques and apply array splittings method using Numpy.

#### Description:
- Perform different forms of slicing.
- Assign index
- Do the different join function, horizontal join, vertical join and depth join

#### Programs:

```python
import numpy as np

arr=np.array([[-1, 2, 0, 4], [4, -05.5, 6, 0], [2.9, 0, 7, 6], [3, -7, 4, 2.0]])
print(arr)
print(arr.shape)
print(arr.ndim)
```

```python
#Returns every other rows:
print("Every other rows:\n", arr[0:3:2])
```

```python
#Return every other element
arr=np.array([1, 2, 3, 4, 5, 6, 7])
print("Original array:", arr)
print("Return every other elements in the array: ", arr[::2])
```

```python
#Slicing array
temp=arr[:2, :3]
print("Array with first 2 rows and three columns is:\n", temp)
```

```python
print(arr[0:3:2])
```

```python
print(arr[0:3, 0:3])
```

```python
print(arr[::2, ::4])
```

```python
arr2=np.array([[[1, 2, 3], [3, 4, 5]]])
print(arr2.ndim)
```

```python
#Integer array indexing
temp=arr[[0, 1, 2, 3], [3, 2, 1, 0]]
print("\nElements at indices (0, 3), (1, 2), (2, 1),""(3, 0):\n", temp)
```

```python
import numpy as np
arr = np.array([[-1, 2, 0, 4], [4, -0.5, 6, 0], [2.6, 0, 7, 8], [3, -7, 4, 2.0]])
```

```python
#Boolean array indexing
cond = arr>2

#cond is a boolean array
temp = arr[cond]
print("Elements greater than 2:\n", temp)
```

```python
#Joining two arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
arr = np.concatenate((arr1, arr2))
print("Original arrays:\n", arr1, arr2) 
print("\nJoined array:\n", arr)
```

```python
#Horizontal Join
arr = np.hstack((arr1, arr2))
print("Horizontal joining:\n", arr)
```

```python
#Vertical join
arr = np.vstack((arr1, arr2))
print("Vertical joining:\n", arr)
```

```python
#Depth join
arr = np.dstack((arr1, arr2))
print(arr)
```

```python
#Splitting of array
arr = np.array ([1, 2, 3, 4, 5, 6])
newarr = np.array_split(arr, 3)
print("Original array:\n", arr)
print("\nSplitted array in another form:")

#Displaying splitted array in another form
print(newarr[0])
print(newarr[1])
print(newarr[2])
```
