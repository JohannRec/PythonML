```python
import pandas as pd
import numpy as np
#Creating Series from Array
arr = np.array(['P','a','n','d','a','s'])
a = pd.Series(arr)
print("Series from array:")
print(a)
```

```python
#Series from dictionary
arr= {'x': 0., 'y': 1., 'z': 2.}
b=pd.Series(arr)
print("\n\nSeries from dictionary:\n")
print (b)
```

```python
#Series using scalar
x= pd.Series(4, index=[0, 1, 2, 3])
print("nSeries using scalar\n")
print (x)
```

```python
#Series using scalar
x = pd.Series(4, index=[0, 1, 2, 3])
print("\nSeries using scalar\n")
print (x)
```
